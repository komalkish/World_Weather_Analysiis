```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
lats = np.random.uniform(low = -90.000, high = 90.000, size = 1500)
langs = np.random.uniform(low = -180.000, high = 180.000, size = 1500)
lat_langs = zip(lats, langs)
lat_langs
coordinates = list(lat_langs)
```

```python
from citipy import citipy
```

```python
for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
        citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
#create list for for holding name of cities
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    #check duplicates
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
# Create an endpoint URL for a city.
city_url = url + "&q=" + "Boston"
city_weather = requests.get(city_url)
city_weather.json()
```

```python
boston_data = city_weather.json()
boston_data["coord"]["lat"]
```

```python
lat = boston_data["coord"]["lat"]
lng = boston_data["coord"]["lon"]
max_temp = boston_data["main"]["temp_max"]
humidity = boston_data["main"]["humidity"]
clouds = boston_data["clouds"]["all"]
wind = boston_data["wind"]["speed"]
print(lat, lng, max_temp, humidity, clouds, wind)
```

```python
citu = ["Danville", "New York", "Seattle", "Hayward"]
```

```python
for city in citu:
    try:
        city_result = url +"&q" + city
        weather = requests.get(city_result).json()
        tempre = weather["main"]["temp"]
    except KeyError:
        print("Key Error received for " + city)
        continue
    print(f"The weather is" +str(tempre)+ "degress in" +city)
```
